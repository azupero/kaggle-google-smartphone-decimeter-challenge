```python
import numpy as np
import pandas as pd
from plotly import express as px
from pathlib import Path

DATA_PATH = "../../input"
```

```python
sub = pd.read_csv(Path(DATA_PATH) / "sample_submission.csv")
sub.head()
```

```python
def visualize_trafic(df: pd.DataFrame, center: dict, zoom: int = 9):
    fig = px.scatter_mapbox(
        df,
        lat="latDeg",
        lon="lngDeg",
        color="phoneName",
        labels="phoneName",
        zoom=zoom,
        center=center,
        height=600,
        width=1200
    )
    fig.update_layout(mapbox_style="stamen-terrain", margin={"r": 0, "t": 0, "l": 0, "b": 0}, title_text="GPS trafic")
    fig.show()
```

```python
df_sample_trail_gt = pd.read_csv(Path(DATA_PATH) / "train/2020-05-14-US-MTV-1/Pixel4/ground_truth.csv")
center = {"lat": 37.423576, "lon": -122.094132}

visualize_trafic(df_sample_trail_gt, center)
```
